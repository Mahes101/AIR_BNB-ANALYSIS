```python
df_countries_geocountries = {
    'Portugal':'Portugal',
 'Brazil':'Brazil',
 'United States':'United States of America',
 'Turkey':'Turkey',
 'Canada':'Canada',
 'Spain':'Spain',
 'Australia':'Australia',
 'China':'China'
}
```

```python
df_teams_countries_price = df.groupby(by="Country").agg({'Price':'count'}).reset_index().sort_values(by='Price', ascending=False)
ax = df_teams_countries_price.plot.bar(x='Country', xlabel = '', figsize=(20,8))
```

```python
for country in df_countries_geocountries.keys():
    print(df[df['Country'] == country]['Country'].tolist()[0])
```

```python
print("Before", list(set(df['Country'].unique().tolist()).intersection(df_countries_geocountries.values())), list(set(df['Country'].unique().tolist()).intersection(df_countries_geocountries.keys())))
for country in df_countries_geocountries.keys():
    df.loc[df['Country'] == country, 'Country'] = df_countries_geocountries[country]
print("After", list(set(df['Country'].unique().tolist()).intersection(df_countries_geocountries.values())), list(set(df['Country'].unique().tolist()).intersection(df_countries_geocountries.keys())))
```

```python
df_world_teams = df_world.merge(df_teams_countries_price, how="left", left_on=['name'], right_on=['Country'])
print("Type of DataFrame : ", type(df_world_teams), df_world_teams.shape[0])
df_world_teams.head()
```
